### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
tiendas = {
    "Tienda 1": tienda1,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

# Calcular facturación total de cada tienda mediante comprensión de diccionario
facturaciones = {nombre_tienda:float(df['Precio'].sum()) for nombre_tienda, df in tiendas.items()}

print('Facturación Total por Tienda')
for tienda, facturacion in facturaciones.items():
    print(f"{tienda}: ${facturacion:,.2f}")

# Encontrar la tienda con menor facturación
tienda_menos_vende = min(facturaciones, key=facturaciones.get)
print(f'\nLa tienda con menor facturación es: {tienda_menos_vende} con ${facturaciones[tienda_menos_vende]:,.2f}')
```

# 2. Ventas por categoría

```python
# Se unirán todos los DataFrames de las 4 tiendas en uno solo, para luego realizar cálculos mediante tabla pivote
# Agregar una columna 'Nombre_Tienda' a cada DataFrame
tienda1['Nombre_Tienda'] = 'Tienda 1'
tienda2['Nombre_Tienda'] = 'Tienda 2'
tienda3['Nombre_Tienda'] = 'Tienda 3'
tienda4['Nombre_Tienda'] = 'Tienda 4'
# Concatenar todos los DataFrames en uno solo
df_consolidado = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
df_consolidado.tail()
```

```python
# Crear la tabla pivote que contiene la cantidad de ventas de cada categoría por tienda
ventas_por_categoria_por_tienda = pd.pivot_table(
    df_consolidado,
    index='Categoría del Producto', # index='Categoría del Producto': Hace que las categorías sean las filas
    columns='Nombre_Tienda',# columns='Nombre_Tienda': Hace que los nombres de las tiendas sean las columnas
    values='Producto',  # Se utiliza cualquier columna que se pueda contar para representar una venta.
    aggfunc='count', # aggfunc='count': La función de agregación es contar el número de productos/ventas.
    fill_value=0 # Rellena los valores NaN (categorías sin ventas en una tienda) con 0
)

print('\nVentas por Categoría y por Tienda')
ventas_por_categoria_por_tienda
```

```python
# Gracias a .idxmax() se puede encontrar el índice (la categoría) de la fila con el valor máximo en cada columna
categoria_mas_popular_por_tienda = ventas_por_categoria_por_tienda.idxmax()

categoria_mas_popular_por_tienda
```

# 3. Calificación promedio de la tienda


```python
df_consolidado['Calificación'].value_counts()
```

```python
# Calcular Calificación Promedio de cada tienda mediante comprensión de diccionario
calificaciones = {nombre_tienda:float(df['Calificación'].mean()) for nombre_tienda, df in tiendas.items()}

print('Calificación promedio por Tienda')
for tienda, calificacion in calificaciones.items():
    print(f"{tienda}: {calificacion:,.2f}")
```

```python
#Identificar la tienda con la calificación más baja
tienda_baja_calificacion = min(calificaciones, key=calificaciones.get)
print(f"\nLa tienda con la calificación promedio más baja es: {tienda_baja_calificacion} con {calificaciones[tienda_baja_calificacion]:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Saber cuántos productos diferentes existen en estas 4 tiendas
df_consolidado['Producto'].nunique()
```

```python
print('Productos más y menos vendidos por tienda')

for nombre_tienda, df_tienda in tiendas.items():
    print(f'\n### {nombre_tienda}')
    # Asegurar que el DataFrame de la tienda actual contenga registros
    if df_tienda.empty:
        print(f'\n{nombre_tienda}: No hay datos de ventas para analizar.')
        continue

    # Contar las ventas de cada producto en el DataFrame de cada tienda
    # .vlue_counts() devuelve una Serie donde el índice será el producto y el valor el número de ventas
    # Por defecto .value_counts() ordena de mayor a menor
    conteo_productos = df_tienda['Producto'].value_counts()
    print(f'Existen {len(conteo_productos)} productos diferentes vendidos en esta tienda')

    # Productos Más Vendidos
    print("\n#### Productos Más Vendidos:")
    if not conteo_productos.empty:
        # Mostrar los 5 más vendidos, o menos si no hay tantos productos diferentes
        top_5 = conteo_productos.head(5)
        # Iterar sobre los 5 primeros de la Serie que devolvió .value_counts()
        for producto, ventas in top_5.items():
                print(f'- {producto}: {ventas} unidades')
    else:
        print("No hay productos vendidos para identificar los más vendidos.")

    # Productos Menos Vendidos
    print('\n#### Productos Menos Vendidos:')
    if len(conteo_productos)>5:
        # Mostrar los 5 productos con menos ventas (al menos 1 venta)
        # value_counts() solo incluye productos que tienen al menos una venta
        # value_counts() ordena de mayor a menor, así que tail(5) da los 5 menos vendidos
        bottom_5 = conteo_productos.tail(5)
         # Iterar sobre los 5 últimos de la Serie que devolvió .value_counts()
        for producto, ventas in bottom_5.items():
            print(f"- {producto}: {ventas} unidades")
    else:
        print("No hay tantos productos diferentes vendidos para identificar los menos vendidos.")
    print("\n" + "=" * 50) # Separador entre tiendas
```

# 5. Envío promedio por tienda

```python
# Calcular envío promedio de cada tienda mediante comprensión de diccionario
envios = {nombre_tienda:float(df['Costo de envío'].mean()) for nombre_tienda, df in tiendas.items()}

print('Costo Promedio de Envíos por Tienda')
for tienda, envio in envios.items():
    print(f"{tienda}: ${envio:,.2f}")

# Encontrar la tienda con mayor costo promedio de envío
tienda_envios_caros = max(envios, key=envios.get)
print(f'\nLa tienda con mayor costo promedio de envío es: {tienda_envios_caros} con ${envios[tienda_envios_caros]:,.2f}')
```

## Generando gráficos con la información calculada

```python
# Facturación Total por Tienda (Barras)
plt.figure(figsize=(8, 5)) # Ajusta el tamaño del gráfico
tiendas_nombres = list(facturaciones.keys())
facturacion_valores = list(facturaciones.values())

bars = plt.bar(tiendas_nombres, facturacion_valores, color=['lightblue', 'pink', 'MediumAquamarine', 'Khaki'])
plt.xlabel("Tienda", fontsize=12)
plt.ylabel("Facturación Total ($)", fontsize=12)
plt.title("Facturación Total por Tienda", fontsize=15)
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.ticklabel_format(style='plain', axis='y') # Evitar notación científica en el eje Y

# Añadir etiquetas de valor en las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01 * max(facturacion_valores), f'${yval:,.2f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# Peso de la facturación de cada tienda
total_ventas= df_consolidado['Precio'].sum()
porcentaje_ventas = {nombre_tienda:round(float(df['Precio'].sum()/total_ventas*100),1) for nombre_tienda, df in tiendas.items()}
porcentaje_ventas
```

```python
# Generación del Gráfico de Torta (Pie Chart)
plt.figure(figsize=(5, 5)) # Ajusta el tamaño del gráfico

    # Datos para el gráfico de torta
labels = porcentaje_ventas.keys()
sizes = porcentaje_ventas.values()
colors = ['lightblue', 'pink', 'MediumAquamarine', 'Khaki'] # Colores suaves para las rebanadas
explode = [0.0 if p < 10 else 0.05 for p in sizes] # Resalta un poco las rebanadas grandes

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
            autopct='%1.1f%%', shadow=True, startangle=140, textprops={'fontsize': 12})

plt.title('Porcentaje de Ventas Representado por Cada Tienda', fontsize=16)
plt.axis('equal') # Asegura que el círculo sea proporcional
plt.tight_layout()
plt.show()
```

```python
# Ventas por Categoría de Producto por Tienda (Barras Apiladas)
# Se hace uso de la tabla pivote directamente para este gráfico
# Transponer para tener categorías en el eje X y tiendas como barras apiladas
ventas_por_categoria_por_tienda.plot(kind='bar', stacked=True, figsize=(12, 6), cmap='viridis')

plt.xlabel("Categoría del Producto", fontsize=12)
plt.ylabel("Número de Ventas", fontsize=12)
plt.title("Distribución de Ventas por Categoría y por Tienda", fontsize=15)
plt.xticks(rotation=45, ha='right') # Rotar etiquetas para mejor lectura
plt.legend(title="Tienda", bbox_to_anchor=(1.05, 1), loc='upper left') # Leyenda fuera del gráfico
plt.grid(axis='y', linestyle='--')
plt.tight_layout() # Ajustar el diseño para que la leyenda no se superponga
plt.show()
```

```python
# Cálculo del precio promedio por categoría en df_consolidado
precios_promedio_por_categoria = df_consolidado.groupby('Categoría del Producto')['Precio'].mean().sort_values(ascending=False)

print("--- Precio Promedio por Categoría (Todas las Tiendas Consolidadas) ---")
print(precios_promedio_por_categoria.to_string()) # Con .to_string() se muestran todas las filas
print("-" * 60)

# --- Visualización (Gráfico de Barras) ---
plt.figure(figsize=(10, 6))
precios_promedio_por_categoria.plot(kind='bar', color='skyblue')
plt.title('Precio Promedio por Categoría de Producto (Todas las Tiendas)', fontsize=15)
plt.xlabel('Categoría del Producto', fontsize=12)
plt.ylabel('Precio Promedio ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='y') # Evitar notación científica en el eje Y
plt.tight_layout()
plt.show()
```

## Análisis de la información obtenida

El presente informe tiene como objetivo brindar un análisis sobre el rendimiento de 4 tiendas, con la finalidad de sustentar la difícil decisión sobre cuál de ellas se debería vender para invertir en un nuevo negocio. Se han evaluado distintos factores clave como ingresos totales, desempeño por categoría de producto, satisfacción de cliente, productos más y menos vendidos y costos promedio de envíos.


**Ingresos Totales (Facturación)**

Tal como se puede visualizar en los gráficos presentados, la Tienda 1 representa la facturación más alta, mientras que la Tienda 4 es la que genera menores ingresos totales, reportando el menor rendimiento en términos de ingresos.

**Ventas por Categoría de Producto**
En el gráfico de barras apiladas puede comprobarse que la Tienda 4, si bien tiene los ingresos más bajos, posee un desempeño consistente en categorías como Artículos para el hogar y Juguetes, llegando incluso a superar a las otras. Sin embargo,en categorías como Electrodomésticos (la categoría de productos más costosos) e Instrumentos musicales, su rendimiento es de los más bajos y eso contribuiría a su menor facturación total.Las tiendas 3 y 4 representan los mayores volúmenes de electrónicos.

**Calificación promedio de clientes**

Las cuatro tiendas poseen calificaciones promedio similares y bastante altas. La Tienda 1 presenta la calificación más baja, pero con una diferencia muy mínima con respecto a las demás tiendas y por tanto representa mayor preocupación. El gráfico de barras muetra que las cuatro tiendas tienen un buen nivel de satisfacción de clientes.

**Productos más y menos vendidos**

La Tiensa 1 tiene alta demanda de productos duraderos y de alto valor como Microondas, TV LED UHD 4K, secadora de ropa y no se centra en productos educativos como Ciencia de datos con Python.
La Tienda 2 posee una demanda de mayor diversidad desde instrumento musicales hasta Iniciando en Programación. Esto sugiere tal vez un mercado más juvenil, donde Mesa de comedor y Juego de mesa no tiene mucha cabida.
La Tienda 3 parece tener éxito con clientes que desean equipar sus hogares con muebles y sets de cocina, pero posee demanda más baja que las otras tiendas en productos infantiles como Mochila y Bloques de construcción.
La Tienda 4 posee una demanda similar a la Tienda 3 en cuanto a muebles y artículos para el hogar, pero posee una las menores ventas en lo que respecta a Armarios y electrodomésticos grandes como Refrigeradores (artículos de alto valor) y eso parece estar afectando su facturación general.

**Costo Promedio de Envío**

La Tienda 1 registra el costo de envío promedio más alto, mientras que la Tienda 4 posee el costo más bajo. Podría verse influenciado por diferencias en la eficacia logística, diferencias en la dispersión, pero también a que la tienda 4 no posee demasiado éxito en la venta de artículos grandes como heladeras y armarios que debido a peso y tamaño son más caros de transportar.

**Conclusión**

Dados todos los factores analizados se recomienda vender la tienda 4, ya que es la que genera los ingresos más bajos y posee bajo volumen de ventas en categorías de mayor valor como Electrodomésticos. Aunque registra los costos de envío más bajo, esto no compensa el menor volumende ingresos que genera. El nivel de satisfacción del cliente, al ser tan parejo, no parece ser un factor decisivo. Las otras tres tiendas muestran una distribución de ventas más robusta en las categorías de mayor valor, constituyendo activos más sólidos para mantener.
